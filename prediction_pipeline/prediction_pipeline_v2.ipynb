# 📊 Прогноз баллов ЕГЭ по истории

Этот ноутбук автоматически рассчитывает ожидаемый балл ЕГЭ для каждого ученика на основе результатов пробных экзаменов.

**Что вы получите на выходе:**
- Файл Excel с прогнозом по каждому ученику
- Ожидаемый диапазон балла (от / до)
- Зону риска: 🔴 риск / 🟡 средняя / 🟢 сильный
- Ожидаемый средний балл по школе
- Презентацию PowerPoint для руководства

---

## 🤖 Как работает pipeline

Каждый запуск — это полноценное исследование данных. Ноутбук автоматически перебирает несколько алгоритмов и выбирает лучший для ваших данных.

1. **Анализ данных** — система изучает объём и качество исторических данных
2. **Перебор алгоритмов** — тестируется до 8 моделей на реальных исторических данных
3. **Выбор лучшей** — победитель определяется по точности зон и точности числа
4. **Прогноз** — лучшая модель применяется к текущему году

При добавлении новых данных система автоматически пересмотрит выбор модели.

## 📂 Требования к формату файлов

Положите 4 файла в папку рядом с ноутбуком:

| Файл | Что содержит |
|------|-------------|
| `годовой_обучение.csv` | Прошлый год, годовой курс — для обучения |
| `полугодовой_обучение.csv` | Прошлый год, полугодовой курс — для обучения |
| `годовой_прогноз.csv` | Этот год, годовой курс — прогнозируем |
| `полугодовой_прогноз.csv` | Этот год, полугодовой курс — прогнозируем |

**Обязательные колонки в файлах 2025:**
`ученик`, `пробник_1` ... `пробник_N`, `егэ`

**Обязательные колонки в файлах 2026:**
`ученик`, `куратор`, `пробник_1` ... `пробник_N`

**Правила:**
- Пустая ячейка в пробнике — ученик не писал, это нормально
- Пустая ячейка в `егэ` — строка пропускается
- Текст в `егэ` («написала», «нет связи») — считается пустым, строка пропускается
- Лишние колонки игнорируются автоматически

## 📋 Как читать файл с результатами

| Колонка | Что значит |
|---------|------------|
| имя | ФИ ученика |
| куратор | ФИ куратора |
| прогноз | Ожидаемый балл ЕГЭ |
| от / до | Вероятный диапазон реального балла |
| зона | Категория уровня подготовки |

**Что означает «—» в колонках прогноз и диапазон:**

Если ученик написал менее 3 пробников — точный прогноз ненадёжен,
поэтому вместо цифры ставится «—». Зона при этом всё равно показывается —
она отражает текущий уровень ученика и помогает куратору понять
кому нужно уделить внимание.

**Зоны:**
- 🔴 **Риск** — прогноз ниже 64. Рекомендуем уделить дополнительное внимание.
- 🟡 **Средняя** — прогноз 64–78. Стандартная траектория подготовки.
- 🟢 **Сильный** — прогноз выше 78. Ученик демонстрирует высокий уровень.

---

## ⚠️ Важно о точности прогноза

Прогноз построен на исторических данных прошлого года и является **ориентиром**, а не гарантией результата. Средняя погрешность составляет около 8 баллов.

Зоны отражают **текущую траекторию** ученика на основе пробных экзаменов. При изменении динамики подготовки прогноз может измениться.

Рекомендуем использовать прогноз как один из инструментов работы куратора — в сочетании с личным наблюдением и оценкой преподавателя.

## Перспективы совершенствования

Сейчас ученик идентифицируется по имени в колонке `ученик`. Это работает, но есть риск при полных тёзках — они попадут в одну строку или создадут путаницу в результатах.

Рекомендуем один из вариантов:
- Писать **полное ФИО** (Иванов Иван Иванович) — снижает риск совпадений
- Добавить колонку `id` с уникальным номером ученика — надёжнее всего

Если в вашей системе уже есть ID учеников — добавьте его в файлы, можно будет доработать. В следующей версии пайплайна сможем использовать его как основной идентификатор.

## Шаг 1. Загрузка данных

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge, Lasso, ElasticNet, HuberRegressor
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.model_selection import LeaveOneOut
from dataclasses import dataclass, field

г   = pd.read_csv('годовой_обучение.csv')
п   = pd.read_csv('полугодовой_обучение.csv')
г26 = pd.read_csv('годовой_прогноз.csv')
п26 = pd.read_csv('полугодовой_прогноз.csv')

# Автопереименование колонки егэ
for df in [г, п]:
    df.rename(columns={'Балл ЕГЭ': 'егэ', 'БАЛЛ ЕГЭ': 'егэ'}, inplace=True)

# Определяем колонки пробников автоматически
пробники_г   = sorted([c for c in г.columns   if c.startswith('пробник_')], key=lambda x: int(x.split('_')[1]))
пробники_п   = sorted([c for c in п.columns   if c.startswith('пробник_')], key=lambda x: int(x.split('_')[1]))
пробники_г26 = sorted([c for c in г26.columns if c.startswith('пробник_')], key=lambda x: int(x.split('_')[1]))
пробники_п26 = sorted([c for c in п26.columns if c.startswith('пробник_')], key=lambda x: int(x.split('_')[1]))

# Чистим пробники
for df, пробники in [(г, пробники_г), (п, пробники_п), (г26, пробники_г26), (п26, пробники_п26)]:
    for col in пробники:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Чистим егэ
for df in [г, п]:
    df['егэ'] = pd.to_numeric(df['егэ'], errors='coerce')

г = г.dropna(subset=['егэ'])
п = п.dropna(subset=['егэ'])

print(f"Годовой 2025:     {len(г)} учеников с баллом ЕГЭ, {len(пробники_г)} пробников")
print(f"Полугодовой 2025: {len(п)} учеников с баллом ЕГЭ, {len(пробники_п)} пробников")
print(f"Годовой 2026:     {len(г26)} учеников, {len(пробники_г26)} пробников")
print(f"Полугодовой 2026: {len(п26)} учеников, {len(пробники_п26)} пробников")
```

## Шаг 2. Подготовка признаков

```python
def make_features(df, cols, global_means=None):
    X = df[cols].astype(float).copy()
    features = pd.DataFrame(index=df.index)
    features['последний'] = X.apply(
        lambda r: r.dropna().iloc[-1] if len(r.dropna()) > 0 else np.nan, axis=1)
    features['среднее_все'] = X.mean(axis=1)
    features['среднее_посл3'] = X.apply(
        lambda r: r.dropna().iloc[-3:].mean() if len(r.dropna()) >= 1 else np.nan, axis=1)
    features['макс'] = X.max(axis=1)
    features['тренд'] = X.apply(
    lambda r: r.dropna().iloc[-3:].mean() - r.dropna().iloc[:3].mean()
    if len(r.dropna()) >= 2 else np.nan, axis=1)
    features['кол_написал'] = X.notna().sum(axis=1)
    if global_means is None:
        global_means = features.mean()
    features['тренд'] = features['тренд'].fillna(0)
    features = features.fillna(global_means)
    return features, global_means

def make_forecast(df_raw, имя_col, пробники_cols, модель, mae, means, n_pad_to=None):
    work = df_raw[[имя_col] + пробники_cols].copy()

    for col in пробники_cols:
        work[col] = pd.to_numeric(work[col], errors='coerce')

    work.columns = ['имя'] + [f'p{i+1}' for i in range(len(пробники_cols))]

    if n_pad_to:
        for i in range(len(пробники_cols) + 1, n_pad_to + 1):
            work[f'p{i}'] = np.nan

    feature_cols = [f'p{i+1}' for i in range(n_pad_to or len(пробники_cols))]
    X, _ = make_features(work, feature_cols, global_means=means)
    прогноз = np.clip(модель.predict(X), 0, 100)

    прогноз_int = прогноз.round(0).astype(int)
    work['зона'] = [
        '🔴 риск' if x < 64 else ('🟡 средняя' if x < 78 else '🟢 сильный')
        for x in прогноз_int
    ]

    p_cols = [c for c in work.columns if c.startswith('p')]
    мало = work[p_cols].notna().sum(axis=1) < 3

    work['прогноз'] = прогноз_int
    work['от'] = (прогноз - mae).round(0).astype(int).clip(0, 100)
    work['до'] = (прогноз + mae).round(0).astype(int).clip(0, 100)

    work['прогноз'] = work['прогноз'].astype(str)
    work['от'] = work['от'].astype(str)
    work['до'] = work['до'].astype(str)

    work.loc[мало, 'прогноз'] = '—'
    work.loc[мало, 'от'] = '—'
    work.loc[мало, 'до'] = '—'

    return work[['имя', 'прогноз', 'от', 'до', 'зона']]

def баллов(n):
    n = int(round(n))
    if 11 <= n % 100 <= 19: return f'{n} баллов'
    rem = n % 10
    if rem == 1: return f'{n} балл'
    elif 2 <= rem <= 4: return f'{n} балла'
    else: return f'{n} баллов'

def учеников(n):
    n = int(n)
    if 11 <= n % 100 <= 19: return f'{n} учеников'
    rem = n % 10
    if rem == 1: return f'{n} ученик'
    elif 2 <= rem <= 4: return f'{n} ученика'
    else: return f'{n} учеников'

def качество_mae(v):
    if v < 6:     return 'высокая точность'
    elif v <= 10: return 'умеренная точность'
    else:         return 'ориентировочная точность'

def качество_зон(v):
    if v > 0.78:   return 'высокая'
    elif v >= 0.64: return 'умеренная'
    else:           return 'ориентировочная'

X_г, means_г = make_features(г, пробники_г)
y_г = г['егэ'].values
X_п, means_п = make_features(п, пробники_п)
y_п = п['егэ'].values
print('Признаки созданы')
```

## Шаг 3. Выбор лучшей модели

```python
@dataclass
class DataProfile:
    n_samples:  int
    target_std: float
    regime:     str = field(init=False)
    def __post_init__(self):
        if   self.n_samples < 30:  self.regime = 'tiny'
        elif self.n_samples < 80:  self.regime = 'small'
        elif self.n_samples < 300: self.regime = 'medium'
        else:                      self.regime = 'large'

CANDIDATES = [
    dict(name='Ridge a=10',   model=Ridge(alpha=10),                                          min_samples=0),
    dict(name='Ridge a=5',    model=Ridge(alpha=5),                                           min_samples=30),
    dict(name='Ridge a=50',   model=Ridge(alpha=50),                                          min_samples=0),
    dict(name='Lasso',        model=Lasso(alpha=1.0, max_iter=5000),                          min_samples=30),
    dict(name='ElasticNet',   model=ElasticNet(alpha=0.5, l1_ratio=0.5, max_iter=5000),       min_samples=40),
    dict(name='Huber',        model=HuberRegressor(epsilon=1.5, max_iter=300),                min_samples=25),
    dict(name='GradBoost',    model=GradientBoostingRegressor(n_estimators=80, max_depth=2,
                                   learning_rate=0.08, random_state=42),                      min_samples=80),
    dict(name='RandomForest', model=RandomForestRegressor(n_estimators=100, max_depth=3,
                                   random_state=42),                                          min_samples=100),
]

def _зона(x):
    if x < 64: return 'риск'
    elif x < 78: return 'средняя'
    else: return 'сильный'

def composite_score(mae, zone_acc):
    return 0.3 * (1 / (1 + mae)) + 0.7 * zone_acc

def run_experiment(X, y, label):
    profile    = DataProfile(n_samples=len(y), target_std=float(np.std(y)))
    loo        = LeaveOneOut()
    candidates = [c for c in CANDIDATES if profile.n_samples >= c['min_samples']]
    results    = []
    for c in candidates:
        y_pred = []
        for tr, te in loo.split(X):
            c['model'].fit(X.iloc[tr], y[tr])
            y_pred.append(c['model'].predict(X.iloc[te])[0])
        y_pred   = np.array(y_pred)
        mae      = mean_absolute_error(y, y_pred)
        zone_acc = np.mean([_зона(a) == _зона(b) for a, b in zip(y, y_pred)])
        score    = composite_score(mae, zone_acc)
        results.append({**c, 'mae': mae, 'zone_acc': zone_acc, 'score': score})
    results.sort(key=lambda r: -r['score'])
    winner   = results[0]
    baseline = next(r for r in results if r['name'] == 'Ridge a=10')
    print(f"\n{'─'*58}")
    print(f'  СРАВНЕНИЕ МОДЕЛЕЙ — {label}')
    print(f"{'─'*58}")
    print(f"  {'модель':<18} {'MAE':>5}  {'зоны':>5}  {'score':>6}  статус")
    print(f"  {'─'*52}")
    for r in results:
        tag = ''
        if r['name'] == winner['name']: tag = '<- победитель'
        if r['name'] == 'Ridge a=10':  tag += '  [бейзлайн]'
        print(f"  {r['name']:<18} {r['mae']:>5.1f}  {r['zone_acc']:>4.0%}  {r['score']:>6.3f}  {tag}")
    warnings = []
    if winner['name'] in {'GradBoost', 'RandomForest'} and profile.regime in ('tiny', 'small'):
        warnings.append(f"  Внимание: {winner['name']} при n={profile.n_samples} — рекомендуем добавить данные")
    if winner['mae'] < 3.0:
        warnings.append(f"  Внимание: MAE={winner['mae']:.1f} подозрительно низко — проверьте данные")
    if winner['mae'] > profile.target_std * 0.8:
        warnings.append(f"  Модель даёт ограниченное улучшение — рекомендуем добавить данные")
    if warnings:
        print()
        for w in warnings: print(w)
    if winner['score'] > baseline['score']:
        статус = 'improved'
        fin    = winner
    else:
        статус = 'baseline'
        fin    = baseline
    mae_diff  = baseline['mae'] - winner['mae']
    zone_diff = winner['zone_acc'] - baseline['zone_acc']
    print(f"\n{'═'*58}")
    print(f'  {label} — итог')
    print(f"{'═'*58}")
    print(f"  Бейзлайн  Ridge a=10: MAE={baseline['mae']:.1f}  зоны={baseline['zone_acc']:.0%}  score={baseline['score']:.3f}")
    print(f"  Победитель {winner['name']:<10}: MAE={winner['mae']:.1f}  зоны={winner['zone_acc']:.0%}  score={winner['score']:.3f}")
    print(f'  Разница: MAE {mae_diff:+.1f}  зоны {zone_diff:+.0%}')
    if статус == 'improved':
        print(f"  Используем {winner['name']} (score выше на {winner['score']-baseline['score']:.3f})")
    else:
        print('  Оставляем Ridge a=10 (улучшений нет)')
    print(f"{'═'*58}")
    fin['model'].fit(X, y)
    return fin, статус, profile

print('Запуск эксперимента — это займёт около минуты...')
fin_г, статус_г, profile_г = run_experiment(X_г, y_г, 'ГОДОВОЙ КУРС')
fin_п, статус_п, profile_п = run_experiment(X_п, y_п, 'ПОЛУГОДОВОЙ КУРС')
print('Модели выбраны')
```

## Шаг 4. Прогноз и сохранение

```python
итог_г = make_forecast(
    г26, 'ученик',
    пробники_cols=пробники_г26,
    модель=fin_г['model'], mae=fin_г['mae'], means=means_г, n_pad_to=len(пробники_г)
)

итог_п = make_forecast(
    п26, 'ученик',
    пробники_cols=пробники_п26,
    модель=fin_п['model'], mae=fin_п['mae'], means=means_п, n_pad_to=len(пробники_п)
)

with pd.ExcelWriter('прогноз.xlsx', engine='openpyxl') as writer:
    итог_г.to_excel(writer, sheet_name='Годовой курс',     index=False)
    итог_п.to_excel(writer, sheet_name='Полугодовой курс', index=False)

средний_балл = int(round(
    pd.to_numeric(итог_г['прогноз'], errors='coerce').mean()
))

def итог_строка(label, fin, статус):
    mae_int  = int(round(fin['mae']))
    zone_pct = int(round(fin['zone_acc'] * 100))
    кач_mae  = качество_mae(fin['mae'])
    кач_зон  = качество_зон(fin['zone_acc'])
    доп = f"улучшенная модель: {fin['name']}" if статус == 'improved' else 'стандартная модель'
    return (f"  {label:<26} ошибка +/-{mae_int} б. ({кач_mae}), "
            f"точность зон {zone_pct}% ({кач_зон}). {доп}.")

print(f"""
{'='*64}
  ПРОГНОЗ ГОТОВ
{'='*64}
  Файл: прогноз.xlsx

  Ожидаемый средний балл по школе (годовой): {средний_балл}

{итог_строка('Годовой курс:', fin_г, статус_г)}
{итог_строка('Полугодовой курс:', fin_п, статус_п)}
""")

print('  Распределение по зонам (годовой):')
for зона, кол in итог_г['зона'].value_counts().items():
    print(f'  {зона}: {учеников(кол)} ({round(кол/len(итог_г)*100)}%)')
print()
print('  Распределение по зонам (полугодовой):')
for зона, кол in итог_п['зона'].value_counts().items():
    print(f'  {зона}: {учеников(кол)} ({round(кол/len(итог_п)*100)}%)')
print()
print('Excel сохранён')
```
